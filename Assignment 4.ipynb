```python
import pandas as pd

airbnb = pd.read_csv("airbnb_listings_usa.csv")
airbnb
```

```python
airbnb = airbnb.set_index(['name','neighbourhood', 'room_type'])
airbnb = airbnb.drop(columns=['Unnamed: 0','id','host_id','host_name', 'neighbourhood_group',  
                              'latitude', 'longitude', 'last_review', 'license', 'state', 'city'])

airbnb = airbnb.dropna()
airbnb
```

```python
#Normalizing price column to be able to be compared with the reviews_per_month column
# airbnb['price'] = airbnb['price'].astype(int)
#airbnb['number_of_reviews'] = airbnb['number_of_reviews']/100
# airbnb

airbnb = (airbnb.subtract(airbnb.min()).divide(airbnb.max()-airbnb.min())*9) + 1
airbnb
```

```python
#Finding K

from sklearn.cluster import KMeans

interia_scores = []
for test_k in range(2, 15, 1):
    print(test_k)
    
    tmp_model = KMeans(n_clusters=test_k)
    tmp_model.fit(airbnb[['price','number_of_reviews']])
    
    score = tmp_model.inertia_
    interia_scores.append((test_k, score))
```

```python
intertia_df = pd.DataFrame(interia_scores, columns=["k", "score"])
intertia_df
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(1,1,1)

intertia_df.plot("k", "score", ax=ax)

ax.set_ylabel("Intertia")

plt.show()
```

```python
#Clustering 

k = 4
model = KMeans(n_clusters=k)
model.fit(airbnb[['price','number_of_reviews']])
airbnb["cluster"] = model.labels_
```

```python
plt.scatter(airbnb['number_of_reviews'], airbnb['price'], c=airbnb["cluster"])
plt.title("KMeans Clustering")
plt.xlabel("Total Rating Count")
plt.ylabel("Price")
```

```python
kmeans = KMeans(n_clusters = 4)
labels = kmeans.fit_predict(airbnb)
labels
```

```python
airbnb_df = pd.DataFrame(labels, index=airbnb.index, columns=['cluster'])
airbnb_df
```
